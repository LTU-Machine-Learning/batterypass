```python
#### eval CELL 1

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
import pandas as pd

# Extract true labels and predictions
true_labels = df_val['true label']
predictions = df_val['gpt0_1']

# Calculate metrics
accuracy = accuracy_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions, pos_label='conformant') # Assuming 'conformant' is the positive class
precision = precision_score(true_labels, predictions, pos_label='conformant')
recall = recall_score(true_labels, predictions, pos_label='conformant')

# Calculate confusion matrix. Define labels for consistent output order.
labels = ['conformant', 'nonconformant']
cm = confusion_matrix(true_labels, predictions, labels=labels)

# Display results
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score (conformant as positive): {f1:.4f}")
print(f"Precision (conformant as positive): {precision:.4f}")
print(f"Recall (conformant as positive): {recall:.4f}")

print("\nConfusion Matrix:")
cm_df = pd.DataFrame(cm, index=['True Conformant', 'True Nonconformant'], columns=['Predicted Conformant', 'Predicted Nonconformant'])
display(cm_df)
```

```python
import numpy as np

# Calculate mean and standard deviation for accuracies
mean_accuracy = np.mean(all_accuracies)
std_accuracy = np.std(all_accuracies)

# Calculate mean and standard deviation for F1 scores
mean_f1_score = np.mean(all_f1_scores)
std_f1_score = np.std(all_f1_scores)

# Calculate mean and standard deviation for precisions
mean_precision = np.mean(all_precisions)
std_precision = np.std(all_precisions)

# Calculate mean and standard deviation for recalls
mean_recall = np.mean(all_recalls)
std_recall = np.std(all_recalls)

print(f"Average Accuracy: {mean_accuracy:.4f}, Std Dev Accuracy: {std_accuracy:.4f}")
print(f"Average F1 Score: {mean_f1_score:.4f}, Std Dev F1 Score: {std_f1_score:.4f}")
print(f"Average Precision: {mean_precision:.4f}, Std Dev Precision: {std_precision:.4f}")
print(f"Average Recall: {mean_recall:.4f}, Std Dev Recall: {std_recall:.4f}")
```
